```python
# imports
import pandas as pd
import numpy as np

# SQL
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Float, inspect
from sqlalchemy.sql import text
```

```python
# create filepath
filepath = "Resources/Airbnb_Open_Data_2022.csv"

# read the CSV file
raw_df_2022 = pd.read_csv(filepath, encoding='latin1', low_memory=False)

#view
raw_df_2022.head()
```

```python
# create filepath
filepath = "Resources/listings_2023.csv"

# read the CSV file
raw_df_2023 = pd.read_csv(filepath, encoding='latin1', low_memory=False)

# view
raw_df_2023.head()
```

```python
# create filepath
filepath = "Resources/listings_2024.csv"

# read the CSV file
raw_df_2024 = pd.read_csv(filepath, encoding='latin1', low_memory=False)

#view
raw_df_2024.head()
```

```python
# join these df's
combined_df_1 = pd.concat([raw_df_2022, raw_df_2023], axis=0)
final_df = pd.concat([combined_df_1, raw_df_2024], axis=0)
final_df.info()
```

```python
# drop columns
final_df.drop(columns=['name', 'host_id', 'host_name', 'last_review', 'license', 'number_of_reviews_ltm'], inplace=True)
```

```python
# check the drop
final_df.info()
```

```python
# adjust data for some of the null values to still be useful
# if no minimum_nights, change the value to 1
final_df['minimum_nights'] = final_df['minimum_nights'].fillna(1)

# if no neighbourhood_group or neighbourhood, change to 'unknown' because the lat and lon will show us later
final_df['neighbourhood_group'] = final_df['neighbourhood_group'].fillna('unknown')
final_df['neighbourhood'] = final_df['neighbourhood'].fillna('unknown')

# if number of reviews has no information, change the value to 0
final_df['number_of_reviews'] = final_df['number_of_reviews'].fillna('0')
final_df['reviews_per_month'] = final_df['reviews_per_month'].fillna('0')

# if there is no review rate number, change value to 0 to indicate no reviews or unknown
final_df['review_rate_number'] = final_df['review_rate_number'].fillna('0')

# if there is no calculated host listings count, change the value to 1
final_df['calculated_host_listings_count'] = final_df['calculated_host_listings_count'].fillna('1')

final_df.info()
```

```python
# take out the availability 365 column for now...
final_df.drop(columns=['availability_365'], inplace=True)
final_df.head()
```

```python
final_df.info()
```

```python
# finally, drop the remaining null values that do not include lat/lon or the price
final_df = final_df.dropna()
final_df.info()
```

```python
# save the final df to a new csv for use
final_df.to_csv("Resources/cleaned_airbnb_data.csv")
```
